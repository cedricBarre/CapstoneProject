**Useful Imports**

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
import csv
import scipy.stats
from scipy import stats
```

**Variable Declaration**

```python
INFO = 1 # For debugging

# Change path to find nifti files on local computer
data = []
with open('scans.csv', newline='') as scanfile:
    reader = csv.DictReader(scanfile)
    for row in reader:
        dataRow = [ row['file_name'], float(row['rg_value']), [], 0, 0, 0, 0 ]
        data.append(dataRow)
```

**Extract Images**

```python
idx = 0
for d in data:
    data[idx][2] = np.asarray(nb.load(d[0]).dataobj)
    idx += 1
```

```python
final_image = original_image[mask_image]
final_image.shape
print(final_image)
```

```python
plt.hist(final_image, bins = 100)
plt.show()
```

```python
# Sample from this Random variable
x0 = final_image

p = stats.rayleigh.fit(x0)
X = stats.rayleigh(*p)

# Adjust Distribution parameters
loc, scale = stats.rayleigh.fit(x0)

# Tabulate over sample range (PDF display):
xl = np.linspace(x0.min(), x0.max(), 1000)

# Display Results:
fig, axe = plt.subplots()
axe.hist(x0, bins = 100, density = 1, label="Final Image ")
axe.plot(xl, stats.rayleigh(scale=scale, loc=loc).pdf(xl), label="Adjusted Distribution")
axe.set_title("Distribution Fit")
axe.legend()
axe.grid()
```
