**Useful Imports**

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
import csv
import plotly.graph_objects as go
import PIL as ThePIL
```

**Variable Declaration**

```python
INFO = 1 # For debugging

# Change path to find nifti files on local computer
data = []
with open('scans.csv', newline='') as scanfile:
    reader = csv.DictReader(scanfile)
    for row in reader:
        dataRow = [ row['file_name'], float(row['rg_value']), [], 0, 0, 0, 0 ]
        data.append(dataRow)
```

**Extract Images**

```python
idx = 0
for d in data:
    data[idx][2] = np.asarray(nb.load(d[0]).dataobj)
    idx += 1
mask = np.asarray(nb.load('/home/cedric/Documents/Capstone/CapstoneProject/output/masks/PDw_mask.nii.gz').dataobj)
```

**Sort data scans**

```python
#Averaging the images with the same receiver gain

#ims = np.array([np.array(im) for im in ims])

img_101 = np.array([np.multiply(data[0][2], mask), np.multiply(data[6][2], mask), np.multiply(data[12][2], mask)])
img_50 = np.array([np.multiply(data[1][2], mask), np.multiply(data[7][2], mask), np.multiply(data[13][2], mask)])
img_64 = np.array([np.multiply(data[2][2], mask), np.multiply(data[8][2], mask), np.multiply(data[14][2], mask)])
img_71 = np.array([np.multiply(data[3][2], mask), np.multiply(data[9][2], mask), np.multiply(data[15][2], mask)])
img_80 = np.array([np.multiply(data[4][2], mask), np.multiply(data[10][2], mask), np.multiply(data[16][2], mask)])
img_90 = np.array([np.multiply(data[5][2], mask), np.multiply(data[11][2], mask), np.multiply(data[17][2], mask)])
```

```python
slice = 45
fig, frame = plt.subplots(2, 3, figsize=(14,5))
z1 = np.rot90(img_101[0, :,slice,:])
pos1 = frame[0, 0].imshow(z1)
frame[0, 0].set_title("PDW scan RG 101 #1")
fig.colorbar(pos1, ax=frame[0, 0])

z2 = np.rot90(img_101[1, :,slice,:])
pos2 = frame[0, 1].imshow(z2)
frame[0, 1].set_title("PDW scan RG 101 #2")
fig.colorbar(pos2, ax=frame[0, 1])

z3 = np.rot90(img_101[2, :,slice,:])
pos3 = frame[0, 2].imshow(z3)
frame[0, 2].set_title("PDW scan RG 101 #3")
fig.colorbar(pos3, ax=frame[0, 2])

z1 = np.rot90(img_50[0, :,slice,:])
pos1 = frame[1, 0].imshow(z1)
frame[1, 0].set_title("PDW scan RG 50.8 #1")
fig.colorbar(pos1, ax=frame[1, 0])

z2 = np.rot90(img_50[1, :,slice,:])
pos2 = frame[1, 1].imshow(z2)
frame[1, 1].set_title("PDW scan RG 50.8 #2")
fig.colorbar(pos2, ax=frame[1, 1])

z3 = np.rot90(img_50[2, :,slice,:])
pos3 = frame[1, 2].imshow(z3)
frame[1, 2].set_title("PDW scan RG 50.8 #3")
fig.colorbar(pos3, ax=frame[1, 2])

fig.suptitle('Plot of slice ' + str(slice) + ' for the scans with RG 101 in the first row and RG 50.8 scan in the second row')
plt.savefig("/home/cedric/Documents/Capstone/MallarPresentation/masked_data_101_and_50_8.jpg")
# plt.show()
```

```python
slice = 45
fig, frame = plt.subplots(2, 3, figsize=(14,5))
z1 = np.rot90(data[0][2][:,slice,:])
pos1 = frame[0, 0].imshow(z1, vmax=1.5e8, vmin=0)
frame[0, 0].set_title("a)")
fig.colorbar(pos1, ax=frame[0, 0])

z2 = np.rot90(data[5][2][:,slice,:])
pos2 = frame[0, 1].imshow(z2, vmax=1.5e8, vmin=0)
frame[0, 1].set_title("b)")
fig.colorbar(pos2, ax=frame[0, 1])

z3 = np.rot90(data[4][2][:,slice,:])
pos3 = frame[0, 2].imshow(z3, vmax=1.5e8, vmin=0)
frame[0, 2].set_title("c)")
fig.colorbar(pos3, ax=frame[0, 2])

z1 = np.rot90(data[3][2][:,slice,:])
pos1 = frame[1, 0].imshow(z1, vmax=1.5e8, vmin=0)
frame[1, 0].set_title("d)")
fig.colorbar(pos1, ax=frame[1, 0])

z2 = np.rot90(data[2][2][:,slice,:])
pos2 = frame[1, 1].imshow(z2, vmax=1.5e8, vmin=0)
frame[1, 1].set_title("e)")
fig.colorbar(pos2, ax=frame[1, 1])

z3 = np.rot90(data[1][2][:,slice,:])
pos3 = frame[1, 2].imshow(z3, vmax=1.5e8, vmin=0)
frame[1, 2].set_title("f)")
fig.colorbar(pos3, ax=frame[1, 2])

plt.savefig("/home/cedric/Documents/Capstone/MidProjectReport/figure3.jpg")
```

```python
ratioed_img_101_to_50 = np.divide(img_101, img_50, out=np.zeros_like(img_101), where=img_50!=0)
ratioed_img_101_to_64 = np.divide(img_101, img_64, out=np.zeros_like(img_101), where=img_64!=0)
ratioed_img_101_to_71 = np.divide(img_101, img_71, out=np.zeros_like(img_101), where=img_71!=0)
ratioed_img_101_to_80 = np.divide(img_101, img_80, out=np.zeros_like(img_101), where=img_80!=0)
ratioed_img_101_to_90 = np.divide(img_101, img_90, out=np.zeros_like(img_101), where=img_90!=0)
```

```python
slice = 45
fig, frame = plt.subplots(1, 3, figsize=(18,5))
z1 = np.rot90(ratioed_img_101_to_50[0, :,slice,:])
pos1 = frame[0].imshow(z1)
frame[0].set_title("PDW scan 101/50.8 ratio #1")
fig.colorbar(pos1, ax=frame[0])

z2 = np.rot90(ratioed_img_101_to_50[1, :,slice,:])
pos2 = frame[1].imshow(z2)
frame[1].set_title("PDW scan 101/50.8 ratio #2")
fig.colorbar(pos2, ax=frame[1])

z3 = np.rot90(ratioed_img_101_to_50[2, :,slice,:])
pos3 = frame[2].imshow(z3)
frame[2].set_title("PDW scan 101/50.8 ratio #3")
fig.colorbar(pos3, ax=frame[2])

fig.suptitle("Plot of slice " + str(slice) + " for the 101/50.8 scan ratios")
plt.savefig("/home/cedric/Documents/Capstone/MallarPresentation/ratioed_masked_data_101_to_50_8.jpg")
# plt.show()
```

**Standard Deviation**

```python
std_101_to_50 = np.std(ratioed_img_101_to_50, axis=0)
std_101_to_64 = np.std(ratioed_img_101_to_64, axis=0)
std_101_to_71 = np.std(ratioed_img_101_to_71, axis=0)
std_101_to_80 = np.std(ratioed_img_101_to_80, axis=0)
std_101_to_90 = np.std(ratioed_img_101_to_90, axis=0)
```

```python
slice = 45
fig, frame = plt.subplots(3, 2, figsize=(14,14))
z1 = np.rot90(std_101_to_50[:,slice,:])
pos1 = frame[0, 0].imshow(z1)
frame[0, 0].set_title('Voxel-wise standard deviation \n for the 101 and 50.8 batch of 3 scans')
fig.colorbar(pos1, ax=frame[0,0])

z2 = np.rot90(std_101_to_64[:,slice,:])
pos2 = frame[1, 0].imshow(z2)
frame[1, 0].set_title('Voxel-wise standard deviation for \n the 101 and 64 batch of 3 scans')
fig.colorbar(pos2, ax=frame[1,0])

z3 = np.rot90(std_101_to_71[:,slice,:])
pos3 = frame[0, 1].imshow(z3)
frame[0, 1].set_title('Voxel-wise standard deviation for \n the 101 and 71.8 batch of 3 scans')
fig.colorbar(pos3, ax=frame[0,1])

z4 = np.rot90(std_101_to_80[:,slice,:])
pos4 = frame[1, 1].imshow(z4)
frame[1, 1].set_title('Voxel-wise standard deviation for \n the 101 and 80.6 batch of 3 scans')
fig.colorbar(pos4, ax=frame[1,1])

z5 = np.rot90(std_101_to_90[:,slice,:])
pos5 = frame[2, 0].imshow(z5)
frame[2, 0].set_title('Voxel-wise standard deviation for \n the 101 and 90.5 batch of 3 scans')
fig.colorbar(pos5, ax=frame[2,0])

fig.suptitle("Plot of slice " + str(slice) + " for the voxel-wise standard deviations across the 3 scans for each RG")
plt.savefig("/home/cedric/Documents/Capstone/MallarPresentation/std_masked_data.jpg")
# plt.show()
```

```python
print("Mean of standard deviation for the ratioed scans of RG 101 to RG 50.8 = {}".format(np.mean(std_101_to_50[mask.astype(bool)])))
print("Mean of standard deviation for the ratioed scans of RG 101 to RG 64 = {}".format(np.mean(std_101_to_64[mask.astype(bool)])))
print("Mean of standard deviation for the ratioed scans of RG 101 to RG 71.8 = {}".format(np.mean(std_101_to_71[mask.astype(bool)])))
print("Mean of standard deviation for the ratioed scans of RG 101 to RG 80.6 = {}".format(np.mean(std_101_to_80[mask.astype(bool)])))
print("Mean of standard deviation for the ratioed scans of RG 101 to RG 90.5 = {}".format(np.mean(std_101_to_90[mask.astype(bool)])))
```

**Euclidean Distance**

```python
ed_101_to_50 = np.sqrt((np.power(np.subtract(ratioed_img_101_to_50[0], ratioed_img_101_to_50[1]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_50[1], ratioed_img_101_to_50[2]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_50[2], ratioed_img_101_to_50[0]), 2)))
ed_101_to_64 = np.sqrt((np.power(np.subtract(ratioed_img_101_to_64[0], ratioed_img_101_to_64[1]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_64[1], ratioed_img_101_to_64[2]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_64[2], ratioed_img_101_to_64[0]), 2)))
ed_101_to_71 = np.sqrt((np.power(np.subtract(ratioed_img_101_to_71[0], ratioed_img_101_to_71[1]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_71[1], ratioed_img_101_to_71[2]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_71[2], ratioed_img_101_to_71[0]), 2)))
ed_101_to_80 = np.sqrt((np.power(np.subtract(ratioed_img_101_to_80[0], ratioed_img_101_to_80[1]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_80[1], ratioed_img_101_to_80[2]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_80[2], ratioed_img_101_to_80[0]), 2)))
ed_101_to_90 = np.sqrt((np.power(np.subtract(ratioed_img_101_to_90[0], ratioed_img_101_to_90[1]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_90[1], ratioed_img_101_to_90[2]), 2) + 
                        np.power(np.subtract(ratioed_img_101_to_90[2], ratioed_img_101_to_90[0]), 2)))
```

```python
slice = 45
fig, frame = plt.subplots(3, 2, figsize=(14,14))
z1 = np.rot90(ed_101_to_50[:,slice,:])
pos1 = frame[0, 0].imshow(z1)
frame[0, 0].set_title('Voxel-wise standard deviation \n for the 101 and 50.8 batch of 3 scans')
fig.colorbar(pos1, ax=frame[0,0])

z2 = np.rot90(ed_101_to_64[:,slice,:])
pos2 = frame[1, 0].imshow(z2)
frame[1, 0].set_title('Voxel-wise standard deviation for \n the 101 and 64 batch of 3 scans')
fig.colorbar(pos2, ax=frame[1,0])

z3 = np.rot90(ed_101_to_71[:,slice,:])
pos3 = frame[0, 1].imshow(z3)
frame[0, 1].set_title('Voxel-wise standard deviation for \n the 101 and 71.8 batch of 3 scans')
fig.colorbar(pos3, ax=frame[0,1])

z4 = np.rot90(ed_101_to_80[:,slice,:])
pos4 = frame[1, 1].imshow(z4)
frame[1, 1].set_title('Voxel-wise standard deviation for \n the 101 and 80.6 batch of 3 scans')
fig.colorbar(pos4, ax=frame[1,1])

z5 = np.rot90(ed_101_to_90[:,slice,:])
pos5 = frame[2, 0].imshow(z5)
frame[2, 0].set_title('Voxel-wise standard deviation for \n the 101 and 90.5 batch of 3 scans')
fig.colorbar(pos5, ax=frame[2,0])

fig.suptitle("Plot of slice " + str(slice) + " for the voxel-wise standard deviations across the 3 scans for each RG")
plt.savefig("/home/cedric/Documents/Capstone/MallarPresentation/std_masked_data.jpg")
# plt.show()
```
