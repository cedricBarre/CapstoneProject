**Useful Imports**

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as col
import csv
```

**Variable Declaration**

```python
INFO = 1 # For debugging

# Change path to find nifti files on local computer
data = []
with open('scans.csv', newline='') as scanfile:
    reader = csv.DictReader(scanfile)
    for row in reader:
        dataRow = [ row['file_name'], float(row['rg_value']), [], 0, 0, 0, 0 ]
        data.append(dataRow)
```

**Extract Images**

```python
scan_rg_101_1 = np.asarray(nb.load(data[0][0]).dataobj)
print(data[0][1])
scan_rg_101_2 = np.asarray(nb.load(data[1][0]).dataobj)
print(data[1][1])
scan_rg_50_1 = np.asarray(nb.load(data[2][0]).dataobj)
print(data[2][1])
scan_rg_50_2 = np.asarray(nb.load(data[3][0]).dataobj)
print(data[3][1])
```

**Average Images**

```python
ratio1 = scan_rg_50_1/scan_rg_101_1
ratio2 = scan_rg_50_2/scan_rg_101_2
std1 = np.std(ratio1)
std2 = np.std(ratio2)
print("std dev 1 = {} and std dev 2 = {}".format(std1, std2))
diff = ratio1 - ratio2
print(diff.shape)
fig = plt.figure(1)
z1 = np.rot90(diff[:,40,:])
pos7 = plt.imshow(z1, vmin=0, vmax=0.2)
plt.colorbar(pos7)
plt.title("Ratio PDW scan of 50/101 receiver gain difference")
plt.show()
fig = plt.figure(2)
z1 = np.rot90(ratio1[:,40,:])
pos7 = plt.imshow(z1, vmin=0, vmax=0.5)
plt.colorbar(pos7)
plt.title("Ratio 1 of PDW scan of 50/101 receiver gain")
plt.show()
fig = plt.figure(3)
z1 = np.rot90(ratio2[:,40,:])
pos7 = plt.imshow(z1, vmin=0, vmax=0.5)
plt.colorbar(pos7)
plt.title("Ratio 2 of PDW scan of 50/101 receiver gain")
plt.show()
```
