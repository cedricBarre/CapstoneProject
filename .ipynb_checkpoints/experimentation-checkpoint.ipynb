**Useful Imports**

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
```

**Variable Declaration**

```python
# Change path to find nifti files on local computer
PDW_RG101_1 = "/home/cedric/Documents/Capstone/Brains/MTR_ratio_data/scans/20220210_MCH_MTR_RG_002_1-02-1-PDw-flash_2019.nii.gz"
PDW_RG101_2 = "/home/cedric/Documents/Capstone/Brains/MTR_ratio_data/scans/20220210_MCH_MTR_RG_002_1-05-1-PDw-flash_2019.nii.gz"
PDW_RG50_8_1 = "/home/cedric/Documents/Capstone/Brains/MTR_ratio_data/scans/20220210_MCH_MTR_RG_002_1-04-1-PDw-flash_2019.nii.gz"
PDW_RG50_8_2 = "/home/cedric/Documents/Capstone/Brains/MTR_ratio_data/scans/20220210_MCH_MTR_RG_002_1-06-1-PDw-flash_2019.nii.gz"
```

**Extract Images**

```python
pdw_rg101_1_obj = nb.load(PDW_RG101_1)
pdw_rg101_2_obj = nb.load(PDW_RG101_2)
pdw_rg50_8_1_obj = nb.load(PDW_RG50_8_1)
pdw_rg50_8_2_obj = nb.load(PDW_RG50_8_2)

pdw_rg101_1_array = np.asarray(pdw_rg101_1_obj.dataobj)
pdw_rg101_2_array = np.asarray(pdw_rg101_2_obj.dataobj)
pdw_rg50_8_1_array = np.asarray(pdw_rg50_8_1_obj.dataobj)
pdw_rg50_8_2_array = np.asarray(pdw_rg50_8_2_obj.dataobj)
```

**Plot Images**

```python
slice = 40 # Adjust this to change slice plotted


fig, frame = plt.subplots(2, 2, figsize=(15,15))
frame[0, 0].imshow(np.rot90(pdw_rg101_1_array[:,slice,:]))
frame[0, 0].set_title("PDW scan with 101 receiver gain #1")

frame[1, 0].imshow(np.rot90(pdw_rg101_2_array[:,slice,:]))
frame[1, 0].set_title("PDW scan with 101 receiver gain #2")

frame[0, 1].imshow(np.rot90(pdw_rg50_8_1_array[:,slice,:]))
frame[0, 1].set_title("PDW scan with 50.8 receiver gain #1")

frame[1, 1].imshow(np.rot90(pdw_rg50_8_2_array[:,slice,:]))
frame[1, 1].set_title("PDW scan with 50.8 receiver gain #2")

fig.suptitle("Plot of slice " + str(slice) + " for the scans used for analysis")
plt.show()
```
