**Useful Imports**

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
import csv
```

**Variable Declaration**

```python
INFO = 1 # For debugging

# Change path to find nifti files on local computer
data = []
with open('scans.csv', newline='') as scanfile:
    reader = csv.DictReader(scanfile)
    for row in reader:
        dataRow = [ row['file_name'], float(row['rg_value']), [], 0, 0, 0, 0 ]
        data.append(dataRow)
```

**Extract Images**

```python
idx = 0
for d in data:
    data[idx][2] = np.asarray(nb.load(d[0]).dataobj)
    idx += 1
```

```python
slice = 40 # Adjust this to change slice plotted

original_image = data[5][2]

z0 = np.rot90(original_image[:,slice,:])
pos0 = plt.imshow(z0)
plt.title("Originl PDW scan with {} receiver gain \n Image 1".format(data[5][1]))
plt.colorbar(pos0)
plt.show()
```

```python
z1 = np.rot90(data[0][2][:,slice,:])
pos1 = plt.imshow(z1)
plt.title("Segmented PDW scan with {} receiver gain \n Image 1".format(data[0][1]))
plt.colorbar(pos1)
plt.show()
```
